```python
# Gerando áudio com sua voz clonada em inglês
tts.tts_to_file(
    text="Hello! This is my cloned voice speaking in English. Pretty amazing, right?",
    speaker_wav="/content/minha_voz.wav",
    language="en",
    file_path="/content/teste_ingles.wav"
)

print("Áudio gerado com sucesso!")
```

```python
from deep_translator import GoogleTranslator

roteiro_pt = "Olá! Hoje vou te mostrar como usar inteligência artificial para criar vídeos em outros idiomas. É muito mais simples do que parece."

# Traduzindo para inglês
roteiro_en = GoogleTranslator(source='pt', target='en').translate(roteiro_pt)

print("Original:", roteiro_pt)
print("Traduzido:", roteiro_en)
```

```python
# Pipeline completo: roteiro em PT → tradução → voz clonada
roteiro_pt = "Olá! Hoje vou te mostrar como usar inteligência artificial para criar vídeos em outros idiomas. É muito mais simples do que parece."

# Traduzindo
roteiro_en = GoogleTranslator(source='pt', target='en').translate(roteiro_pt)
print("Traduzido:", roteiro_en)

# Gerando com sua voz clonada
tts.tts_to_file(
    text=roteiro_en,
    speaker_wav="/content/minha_voz.wav",
    language="en",
    file_path="/content/drive/MyDrive/video-clone-pipeline/roteiro_ingles.wav"
)

print("✅ Áudio do roteiro gerado!")
```
